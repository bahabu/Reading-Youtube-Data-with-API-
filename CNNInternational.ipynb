```python
video_ids_auto = ""
for i in range(len(videodata['items'])):
    a = videodata["items"][i]["id"]['videoId']
    video_ids_auto += a+"," 
video_ids_auto = video_ids_auto[:-1]

video_ids_auto
```

```python
url = f"https://www.googleapis.com/youtube/v3/videos?part=snippet,contentDetails,statistics&id={video_ids_auto}&key={api_key}"
json_url = urllib.request.urlopen(url)
video_stats_data = json.loads(json_url.read())
video_stats_data
```

```python
item = []
for i in range(len(video_stats_data['items'])):
    a = video_stats_data["items"][i]["statistics"]
    item.append(a)
print(item)
```

```python
fields = ["viewCount", "likeCount", "dislikeCount", "favoriteCount", "commentCount"]

with open("youtuberstat.csv", "w", newline="") as f:
    writer = csv.DictWriter(f, fieldnames = fields)
     
    writer.writeheader() 
    
    writer.writerows(item)
```

```python
add_list = []
def add_col(add):
    for i in range(50):
        b = video_stats_data["items"][i]["snippet"][add]
        add_list.append(b)
    return add_list
```

```python
dislike_mean_corona = stat.mean(df_corona["dislikeCount"])
like_mean_corona = stat.mean(df_corona["likeCount"])
rating_mean_corona = like_mean_corona/dislike_mean_corona

dislike_mean = stat.mean(df["dislikeCount"])
like_mean = stat.mean(df["likeCount"])
rating_mean = like_mean/dislike_mean
```

```python
df_corona_plot = pd.DataFrame({
    'Factor': ['Non-corona','Corona'],
    'rating_mean': [rating_mean, rating_mean_corona]
})
tidy = df_corona_plot.melt(id_vars='Factor').rename(columns=str.title)
```

```python
fig, ax1 = plt.subplots(figsize=(10,5))
sns.barplot(x='Factor', y='Value', data=tidy, ax=ax1).set(ylabel='Like/dislike ratio',xlabel='')
sns.despine(fig)
```
